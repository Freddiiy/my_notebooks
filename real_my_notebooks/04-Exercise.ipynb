````python
import numpy as np
import matplotlib.pyplot as plt

neighb = {1: 'Indre By', 2: 'Østerbro', 3: 'Nørrebro', 4: 'Vesterbro/Kgs. Enghave',
          5: 'Valby', 6: 'Vanløse', 7: 'Brønshøj-Husum', 8: 'Bispebjerg', 9: 'Amager Øst',
          10: 'Amager Vest', 99: 'Udenfor'}

with open("../data/befkbhalderstatkode.csv", "r") as file_obj:
    bef_data = np.genfromtxt(file_obj, delimiter=",", dtype=np.uint, skip_header=1)

bef_data_mask = (bef_data[:, 0] == 2015)
bef_data_2015 = bef_data[bef_data_mask]

calc_sum_pup = {}
for n in neighb:
    pup_list_mask = (bef_data_2015[:, 1] == n)
    pup_list = bef_data_2015[pup_list_mask]

    pup_sum = np.sum(pup_list[:, 1])
    print(f"Population in {neighb[n]}: {pup_sum}")

    calc_sum_pup.update({neighb[n]: pup_sum})

city = calc_sum_pup.keys()
population = calc_sum_pup.values()

plt.bar(city, population, color = "green", width=0.5)

plt.xlabel("Cities")
plt.ylabel("Population")
plt.title("Population across cities in 2015")
plt.show()
````

````python
copy_bef_data = np.copy(bef_data_2015)
cph_mask = (copy_bef_data[:, 1] == 1)
copy_bef_data_cph = copy_bef_data[cph_mask]
over_65_in_cph_mask = (copy_bef_data_cph[:, 2] > 65)
over_65_in_cph = copy_bef_data_cph[over_65_in_cph_mask]

plt.bar(over_65_in_cph[:, 3], over_65_in_cph[:,2], color="red", width=0.5)
plt.title("People over 65 in CPH")
plt.show()
````
